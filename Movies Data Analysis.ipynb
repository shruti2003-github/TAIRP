## Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

## Loading the datatset

```python
df=pd.read_csv("IMDb Movies India Data.csv",encoding="ISO-8859-1")
df
```

## Basic Information of Data

```python
df.info()
```

```python
df.shape
```

## Checking Null Values

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df["Votes"] = df["Votes"].str.replace(",","")
df["Votes"] = df["Votes"].astype(float)
```

```python
df["Duration"] = df["Duration"].str.replace("min","")
df["Duration"] = df["Duration"].astype(float)
```

```python
df.rename(columns={'Duration' : 'Duration(min)'}, inplace=True)
```

```python
df.describe()
```

```python
df.describe(include='O')
```

## DATA ANALYSIS :

```python
# Display the records where the duration of the Movies is exactly 200 minutes.

df.loc[(df["Duration(min)"]==200)]
```

```python
# Average  Ratings:

df['Rating'].mean()
```

```python
# Display the Movie which has high Votes.

Max_vote = df['Votes'].max()
df.loc[(df["Votes"]==Max_vote)]
```

```python
# Find The Movie details in which Akshay Kumar & Katrina Kaif worked together. 

df.loc[(df["Actor 1"]=='Akshay Kumar') & (df["Actor 2"]=='Katrina Kaif')]
```

```python
# Display movies directed by Mahesh Bhatt received more than 1000 votes.

df.loc[(df['Director']=="Mahesh Bhatt") & (df['Votes']>=1000)]
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['Name'][0:20], y=df['Duration(min)'][0:20], data=df)
plt.title('Movies Name')
plt.ylabel('Movies Duration in minutes')
plt.xticks(rotation=90) 
plt.show()

print(("*")*125)
print("\t\t\t\t\t\t**************ANALYSIS***************\n\n~~  The plot displays the duration of the first 20 movies in the dataset. It indicates that the movie durations vary widely, with some being significantly longer or shorter than others.") 
print(("*")*125)
```

```python
# show Top 10 Actors:

top_actors = pd.concat([df['Actor 1'], df['Actor 2'], df['Actor 3']]).value_counts().head(10)
print(top_actors)
```

```python
# Find Maximum and Minimum Ratings:

Max = df['Rating'].max()
print(f'Max Rating:{Max}')

Min = df["Rating"].min()
print(f"Min Rating:{Min}")
```

```python
# Display High rated Movie: 

df.loc[(df["Rating"]==Max)]
```

```python
# Display lowest rated Movie: 

df.loc[(df["Rating"]==Min)]
```

```python
# Top 10 Highly Voted Movies: 

Top_10=df.sort_values("Votes",ascending=False).head(10)
Top_10
```

```python
plt.figure(figsize=(6,6))
sns.distplot(df['Votes'],hist=False)
plt.show()

print(("*")*125)
print("\t\t\t\t\t\t**************ANALYSIS***************\n\n~~  The distribution plot shows that the majority of movies in the dataset have a relatively low number of votes, with a peak around the lower end of the vote count. This suggests that while some movies may receive a high number of votes, the majority have fewer votes, indicating varying levels of popularity among the movies in the dataset..") 
print(("*")*125)
```

```python
# Find the Movies names, rates and votes.

df[['Name','Genre','Rating','Votes','Director']].loc[(df['Genre']=='Horror, Mystery, Thriller')]
```

```python
# Identify movies directed by directors whose names contain "Dhawan" and have a rating of 3 or lower.

df[['Name','Director','Actor 1','Actor 2','Actor 3']][df['Director'].str.contains('Dhawan',case=False)].loc[(df['Rating']<=3)]
```

```python
print(df[['Rating','Votes']])

plt.figure(figsize=(8, 6))
plt.scatter(x=df['Rating'][0:20], y=df['Votes'][0:20])
plt.title('Rating vs. Votes')
plt.xlabel('Rating')
plt.ylabel('Votes')
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.grid()
plt.show()

print(("*")*125)
print("\t\t\t\t\t\t**************ANALYSIS***************\n\n~~  The scatter plot illustrates the relationship between the ratings and votes for the first 20 movies in the dataset. It appears that there is no clear correlation between the two variables based on this visualization. However, further analysis may reveal patterns or trends in the entire dataset.") 
print(("*")*125)
```

```python
# Top 5 Movies Director has high rates.

top_directors = df.groupby('Director')['Rating'].mean().sort_values(ascending=False).head(5)
print(top_directors)
```

```python
plt.figure(figsize=(6, 5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation by Heatmap')
plt.show()

print(("*")*125)
print("\t\t\t\t\t\t**************ANALYSIS***************\n\n~~  The heatmap visualizes the correlation between different numerical columns in the dataset. The annotation of correlation coefficients helps to interpret the strength and direction of relationships between variables. Stronger correlations are indicated by darker shades, while lighter shades represent weaker correlations.") 
print(("*")*125)
```

```python
sns.pairplot(df)
plt.show()

print(("*")*125)
print("\t\t\t\t\t\t**************ANALYSIS***************\n\n~~  This Visualization shows that Collection of Charts which represents the relation between Variables of data, trends and patterns as well.") 
print(("*")*125)
```

```python
plt.figure(figsize=(6,8))
sns.countplot(data=df,y=df['Genre'][0:30])
plt.xticks(rotation=90)
plt.show()

print(("*")*125)
print("\t\t\t\t\t\t**************ANALYSIS***************\n\n~~  The count plot provides the distribution of movie genres in the dataset. It visualizes the frequency of each genre, allowing for easy comparison and identification of the most prevalent genres. Here, we can observe the popularity of different genres and their relative representations.") 
print(("*")*125)
```
